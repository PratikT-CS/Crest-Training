```python
import pandas as pd
```

```python
df1 = pd.DataFrame({
    'ID': [1, 2, 3],
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
})

df2 = pd.DataFrame({
    'ID': [2, 3, 4],
    'Country': ['USA', 'Canada', 'UK']
})

# Merging based on common column "ID"
result = pd.merge(df1, df2, on='ID')
result
```

```python
df2
```

```python
left_join = pd.merge(df1, df2, on='ID', how='left')
left_join
```

```python
right_join = pd.merge(df1, df2, on='ID', how='right')
right_join
```

```python
outer_join = pd.merge(df1, df2, on='ID', how='outer')
outer_join
```

```python
df1 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Charlie'],
    'Age': [25, 30, 35]
}, index=[1, 2, 3])

df1
```

```python
df2 = pd.DataFrame({
    'Country': ['USA', 'Canada', 'UK']
}, index=[2, 3, 4])

df2
```

```python
# Join based on the index
result = df1.join(df2)
result
```

# Write a  Pandas program to join two dataframes using keys from right dataframe only.

```python
data1 = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                     'P': ['P0', 'P1', 'P2', 'P3'],
                     'Q': ['Q0', 'Q1', 'Q2', 'Q3']}) 

data1
```

```python
data2 = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                      'key2': ['K0', 'K0', 'K0', 'K0'],
                      'R': ['R0', 'R1', 'R2', 'R3'],
                      'S': ['S0', 'S1', 'S2', 'S3']})

data2
```

```python
merged_data = pd.merge(data1, data2, how='right', on=['key1', 'key2'])
merged_data
```

```python
merged_data = pd.merge(data2, data1, how='right', on=['key1', 'key2'])
merged_data
```
