```python
import pandas as pd
```

## Columnwise Data Entry

```python
df = pd.DataFrame(
{"a" : [4 ,5, 6],
"b" : [7, 8, 9],
"c" : [10, 11, 12]},
index = [1, 2, 3])
```

```python
df
```

## Rowise Data Entry

```python
df = pd.DataFrame(
[[4 ,5, 6],
[7, 8, 9],
[10, 11, 12]],
index=[1, 2, 3],
columns=['a', 'b', 'c'])
```

```python
df
```
