```python
import pandas as pd
```

```python
df = pd.DataFrame({
    'City': ['New York', 'New York', 'Chicago', 'Chicago', 'Los Angeles', 'Los Angeles'],
    'Temperature': [32, 28, 25, 22, 30, 29],
    'Humidity': [65, 68, 75, 78, 60, 62]
})

# Grouping by 'City' and calculating the mean temperature and humidity
grouped = df.groupby('City').mean()

grouped
```

```python
df = pd.DataFrame({
    'City': ['New York', 'New York', 'Chicago', 'Chicago', 'Los Angeles', 'Los Angeles'],
    'Temperature': [32, 28, 25, 22, 30, 29],
    'Humidity': [65, 68, 75, 78, 60, 62]
})

# Group by 'City' and apply multiple aggregation functions
grouped = df.groupby('City').agg({
    'Temperature': ['mean', 'max', 'min'],
    'Humidity': ['mean', 'sum']
})
grouped
```

# Write a  Pandas program to split a dataset to group by two columns and then sort the aggregated results within the groups.

```python
df = pd.DataFrame({
'ord_no':[70001,70009,70002,70004,70007,70005,70008,70010,70003,70012,70011,70013],
'purch_amt':[150.5,270.65,65.26,110.5,948.5,2400.6,5760,1983.43,2480.4,250.45, 75.29,3045.6],
'ord_date': ['2012-10-05','2012-09-10','2012-10-05','2012-08-17','2012-09-10','2012-07-27','2012-09-10','2012-10-10','2012-10-10','2012-06-27','2012-08-17','2012-04-25'],
'customer_id':[3001,3001,3005,3001,3005,3001,3005,3001,3005,3001,3005,3005],
'salesman_id': [5002,5005,5001,5003,5002,5001,5001,5006,5003,5002,5007,5001]})
df
```

```python
df_sorted = df.groupby(['customer_id', 'salesman_id']).agg({"purch_amt" : "sum"})
df_sorted
```

```python
df_sorted = df_sorted.sort_values("purch_amt")
df_sorted
```

```python
result = df_sorted['purch_amt'].groupby(level=0, group_keys=False)
result.nlargest()
```
